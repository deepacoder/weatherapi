```python
city = 'Berlin'
#def get_weather(city):
today = str(datetime.now().date() + timedelta(days=1))
day_after = str(datetime.now().date() + timedelta(days=2))
url = "http://api.openweathermap.org/data/2.5/forecast?q="+city+"&appid="+API_key+"&units=metric"
response = requests.request("GET", url)
weather = response.json()
weather
df = pd.json_normalize(weather,['list'],[['city', 'name'],['city','country']], errors='ignore')
df1 = pd.json_normalize(data= weather['list'], record_path = ['weather'])
df1
weather_df = pd.concat([df, df1], axis=1)
weather_df.columns = (weather_df.columns.str.strip().str.replace('.', '_'))
weather_df['dt_txt'] = pd.to_datetime(weather_df['dt_txt'])
weather_df = weather_df[['dt_txt', 'pop','main_temp','wind_speed','city_name', 'city_country','main', 'description']]
weather_df = weather_df.loc[(weather_df['dt_txt'] >= today ) & (weather_df['dt_txt'] < day_after)]
weather_df.rename(columns={'dt_txt' : 'time', 'pop':'chance_of_rain','main_temp':'avg_temp', 'wind_speed': 'wind_speed', 'city_name': 'city_name',	'city_country':'country_code',
                    'main': 'cloud_cover', 'description': 'forecast'})
weather_df['ingested_time'] = datetime.now()
#return weather_df
```
